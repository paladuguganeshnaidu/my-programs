```python
import socket
domain =input("Enter the domain name: ")
ip = socket.gethostbyname(domain)
print(f"IP address of {domain}: {ip}")
```

```python
import nmap

# Create a PortScanner instance
nm = nmap.PortScanner()

# Scan localhost for ports 22-443
nm.scan('98.130.140.62', '22-443')

# Check scan status
print(f"Command line: {nm.command_line()}")

# Iterate over discovered hosts
for host in nm.all_hosts():
    print(f"Host: {host} ({nm[host].hostname()})")
    print(f"State: {nm[host].state()}")

    # Iterate over protocols (tcp, udp)
    for proto in nm[host].all_protocols():
        print(f"Protocol: {proto}")

        # Get ports for this protocol
        ports = nm[host][proto].keys()
        for port in sorted(ports): 
            state = nm[host][proto][port]['state']
            print(f"  Port {port}: {state}") 
```
